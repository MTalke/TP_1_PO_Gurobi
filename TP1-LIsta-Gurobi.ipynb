1) Um sapateiro faz 6 sapatos por hora, se fizer somente sapatos, e 5 cintos por hora, se fizer
somente cintos. Ele gasta 2 unidades de couro para fabricar 1 unidade de sapato e 1 unidade
couro para fabricar uma unidade de cinto. Sabendo-se que o total disponível de couro é de 6
unidades e que o lucro unitário por sapato é de 5 unidades monetárias e o do cinto é de 2
unidades monetárias, pede-se: o modelo do sistema de produção do sapateiro, se o objetivo é
maximizar seu lucro por hora.


```python
import gurobipy as gp
from gurobipy import GRB

try:

    # Create a new model
    m = gp.Model("questao1")

    # Create variables
    x = m.addVar(vtype=GRB.BINARY, name="x")
    y = m.addVar(vtype=GRB.BINARY, name="y")

    # Set objective
    m.setObjective(5 * x + 2 * y, GRB.MAXIMIZE)

    # Add constraint: 2x + y <= 6
    m.addConstr(2 * x + y <= 6, "primeira restricao")

    # Add constraint: 10x + 12y <= 60
    m.addConstr(10 * x + 12 * y <= 60, "segunda restricao")

    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.varName, v.x))

    print('Objetivo: %g' % m.objVal)

except gp.GurobiError as e:
    print('Erro no codigo ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encountered an attribute error')
```

2) Certa empresa fabrica 2 produtos P1 e P2. O lucro por unidade de P1 é de 100 u.m. e o lucro
unitário de P2 é de 150 u.m. A empresa necessita de 2 horas para fabricar uma unidade de P1
e 3 horas para fabricar uma unidade de P2. O tempo mensal disponível para essas atividades é
de 120 horas. As demandas esperadas para os 2 produtos levaram a empresa a decidir que os
montantes produzidos de P1 e P2 não devem ultrapassar 40 unidades de P1 e 30 unidades de
P2 por mês. Construa o modelo do sistema de produção mensal com o objetivo de maximizar o
lucro da empresa.

```python
import gurobipy as gp
from gurobipy import GRB

try:

    # Create a new model
    m = gp.Model("questao2")

    # Create variables
    x = m.addVar(vtype=GRB.BINARY, name="Quantidade do produto 1:")
    y = m.addVar(vtype=GRB.BINARY, name="Quantidade do produto 2:")

    # Set objective
    m.setObjective(100 * x + 150 * y, GRB.MAXIMIZE)
    
    # Add onstraint: x1 + x2 <= 120
    m.addConstr(2 * x + 3 * y <= 120, "primeira restricao")
    
    # Add constraint: x1 <= 40
    m.addConstr(1 * x <= 40, "segunda restricao")

    # Add constraint: x2 <= 30
    m.addConstr(1 * y <= 30, "terceira restricao")

    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.varName, v.x))

    print('Objetivo: %g' % m.objVal)

except gp.GurobiError as e:
    print('Erro no codigo ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encountered an attribute error')
```

3) Uma rede de televisão local tem o seguinte problema: foi descoberto que o programa “A"
com 20 minutos de música e 1 minuto de propaganda chama a atenção de 30.000
telespectadores, enquanto o programa "B", com 10 minutos de música e 1 minuto de
propaganda chama a atenção de 10.000 telespectadores. No decorrer de uma semana, o
patrocinador insiste no uso de no mínimo, 5 minutos para sua propaganda e que não há verba
para mais de 80 minutos de música. Quantas vezes por semana cada programa deve ser levado
ao ar para obter o número máximo de telespectadores? Construa o modelo do sistema.

```python
import gurobipy as gp
from gurobipy import GRB

try:

    # Create a new model
    m = gp.Model("questao3")

    # Create variables
    x = m.addVar(vtype=GRB.BINARY, name="Quantidade de programa A")
    y = m.addVar(vtype=GRB.BINARY, name="Quantidade de Programa B")

    # Set objective
    m.setObjective(30.000 * x + 10.000 * y, GRB.MAXIMIZE)

    # Add constraint: 20x1 + 10x2 <= 80
    m.addConstr(x * 20 + y * 10 <= 80, "primeira restricao")

    # Add constraint: 1x1 + 1x2 >= 5
    m.addConstr(x * 1 + y * 1 >= 5, "segunda restricao")

    
    m.addConstr(x >= 0, "terceira restricao")

    m.addConstr(y >= 0, "quarta restricao")
    
    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.varName, v.x))

    print('Objetivo: %g' % m.objVal)

except gp.GurobiError as e:
    print('Erro no codigo ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encontrado atributo errado')
```

4) Um empresa fabrica 2 modelos de cintos de couro. O modelo M1, de melhor qualidade,
requer o dobro do tempo de fabricação em relação ao modelo M2. Se todos os cintos fossem
do modelo M2, a empresa poderia produzir 1.000 unidades por dia. A disponibilidade de couro
permite fabricar 800 cintos de ambos os modelos por dia. Os cintos empregam fivelas
diferentes, cuja disponibilidade diária é de 400 para M1 e 700 para M2. Os lucros unitários são

de $ 4,00 para M1 e $ 3,00 para M2. Qual o programa ótimo de produção que maximiza o
lucro total diário da empresa? Construa o modelo do sistema descrito.


```python
import gurobipy as gp
from gurobipy import GRB

try:

    # Create a new model
    m = gp.Model("questao4")

    # Create variables
    x = m.addVar(vtype=GRB.BINARY, name="Quantidade de cinto M1")
    y = m.addVar(vtype=GRB.BINARY, name="Quantidade de cinto M2")

    # Set objective
    m.setObjective(4 * x + 3 * y, GRB.MAXIMIZE)

    # Add constraint: x1 + x2 <= 800
    m.addConstr(x + y <= 800, "primeira restricao")

    # Add constraint: x1 <= 400
    m.addConstr(x <= 400, "segunda restricao")
    
    # Add constraint: x2 <= 700
    m.addConstr(y <= 700, "terceira restricao")

    # Add constraint: 2x1 + x2 <= 1000
    m.addConstr(2 * x + y <= 1000, "quarta restricao")

    m.addConstr(y >= 0, "quinta restricao")
    
    m.addConstr(x >= 0, "sexta restricao")

    
    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.varName, v.x))

    print('Objetivo: %g' % m.objVal)

except gp.GurobiError as e:
    print('Erro no codigo ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encontrado atributo errado')
```

5) Uma empresa, após um processo de racionalização de produção, ficou com disponibilidade
de 3 recursos produtivos, R1, R2 e R3. Um estudo sobre o uso desses recursos indicou a
possibilidade de se fabricar 2 produtos P1 e P2. Levantando os custos e consultando o
departamento de vendas sobre o preço de colocação no mercado, verificou-se que P1 daria
um lucro de $ 120,00 por unidade e P2, $ 150,00 por unidade. O departamento de produção
forneceu a seguinte tabela de uso de recursos. Que produção mensal de P1 e P2 traz o maior lucro para a empresa? Construa o modelo do
sistema.

```python
import gurobipy as gp
from gurobipy import GRB

try:

    # Create a new model
    m = gp.Model("questao5")

    # Create variables
    x1r1 = m.addVar(vtype=GRB.BINARY, name="Quantidade de P1 com recurso 1")
    x1r2 = m.addVar(vtype=GRB.BINARY, name="Quantidade de P1 com recurso 2")
    x1r3 = m.addVar(vtype=GRB.BINARY, name="Quantidade de P1 com recurso 3")
    x2r1 = m.addVar(vtype=GRB.BINARY, name="Quantidade de P2 com recurso 1")
    x2r2 = m.addVar(vtype=GRB.BINARY, name="Quantidade de P2 com recurso 2")
    x2r3 = m.addVar(vtype=GRB.BINARY, name="Quantidade de P2 com recurso 3")


    # Set objective
    m.setObjective(120 * (x1r1 + x1r2 + x1r3 ) + 150 * (x2r1 + x2r2 + x2r3 ), GRB.MAXIMIZE)

    # Add constraint: 2.x1r1 + 4.x2r1 <= 100
    m.addConstr(2 * x1r1 + 4 * x2r1 <= 100, "primeira restricao")

    # Add constraint: 3.x1r1 + 2.x2r1 <= 90
    m.addConstr(3 * x1r2 + 2 * x2r2 <= 90, "segunda restricao")
    
    # Add constraint: x2 <= 700
    m.addConstr(5 * x1r3 + 3 * x2r3 <= 120, "terceira restricao")

    m.addConstr(x1r1 >= 0, "quarta restricao")
    
    m.addConstr(x1r2 >= 0, "quinta restricao")

    m.addConstr(x1r3 >= 0, "sexta restricao")

    m.addConstr(x2r1 >= 0, "setima restricao")

    m.addConstr(x2r2 >= 0, "oitava restricao")

    m.addConstr(x2r3 >= 0, "nona restricao")

    
    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.varName, v.x))

    print('Objetivo: %g' % m.objVal)

except gp.GurobiError as e:
    print('Erro no codigo ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encontrado atributo errado')
```
